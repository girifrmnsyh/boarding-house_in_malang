```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
# path set up
PATH = 'C:\chromedriver-win64\chromedriver.exe'
service = Service(executable_path=PATH)
driver = webdriver.Chrome(service=service)

# open page
url = 'https://mamikos.com/kost/kost-malang-murah?gender=0,1&rent=2&sort=price,-&price=10000-1030000&singgahsini=0'
driver.get(url)
driver.maximize_window()
time.sleep(3)
```

```python
# list for data
names = []
locations = []
facilities = []
prices = []
paid_types = []

# get elements
for page in range(1,67):
    # list items
    items = driver.find_elements(by='xpath',value='//div[@class="room-list__card col-12"]')
    print(page)
    
    # get and insert
    for item in items:
        # get data
        name = item.find_element(by='xpath', value='.//span[contains(@class,"rc-info__name")]').text
        location = item.find_element(by='xpath', value='.//span[contains(@class, "rc-info__location")]').text
        facility = item.find_element(by='xpath', value='.//div[contains(@class,"rc-facilities rc")]').text
        price = item.find_element(by='xpath', value='.//span[contains(@class, "rc-price__text")]').text
        paid_type = item.find_element(by='xpath', value='.//span[contains(@class, "rc-price__type")]').text
        
        # insert to list
        names.append(name)
        locations.append(location)
        facilities.append(facility)
        prices.append(price)
        paid_types.append(paid_type)
    
    # next page
    try:
        driver.find_element(by='xpath', value='//a[text()=">"]').click()
        if page < 5:
            time.sleep(5)
        else:
            time.sleep(3)
    except:
        break
```

```python
# dataframe
df = pd.DataFrame({
    'Nama Kos' : names,
    'Lokasi' : locations,
    'Fasilitas' : facilities,
    'Harga' : prices,
    'Peride Bayar' : paid_types
})
df.head()
```
